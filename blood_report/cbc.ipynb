```python
import numpy as np
import pandas as pd
```

```python
df= pd.read_excel("cbc information.xlsx")
```

```python
df.head()
```

```python
medicine_rules = {
    "Hemoglobin": ("<", 12, ["Iron supplements", "Folic Acid", "Vitamin B12"]),
    "RBC": ("<", 4.5, ["Iron", "Folic Acid", "Vitamin B12"]),
    "WBC": ("<", 4000, ["Immune boosters",]),
    "Platelets": ("<", 150000, ["Vitamin B12", "Platelet boosters",]),
    "Hematocrit": ("<", 36, ["Iron supplements", "Hydration"]),
    "MCV": ("<", 80, ["Iron"]),
    "MCH": ("<", 27, ["Iron supplements"]),
    "MCHC": ("<", 32, ["Iron", "Vitamin C"]),
    "RDW": ("<", 11.5, ["Iron", "B12", "Folic Acid"]),
    "Neutrophils": ("<", 40, ["Vitamin C", "Zinc"]),
    "Lymphocytes": ("<", 20, ["Immune boosters", "Vitamin C"]),
    "Monocytes": ("<", 2, ["Monitor", "Check for infection"]),
    "Eosinophils": ("<", 1, ["Consult if <1%"]),
    "Basophils": ("<", 0.5, ["Rarely treated"]),
    "ESR": ("<", 0, ["Investigate inflammation"]),
    "CRP": ("<", 0.5, ["Check inflammation"]),
    "Uric Acid": ("<", 3.5, ["Uric acid diet", "Hydration"]),
    "Creatinine": ("<", 0.6, ["Hydration", "Renal diet"]),
    "Urea": ("<", 15, ["Hydration", "Reduce protein intake"]),
    "Bilirubin": ("<", 0.3, ["Liver-supporting meds"]),
    "SGPT/ALT": ("<", 10, ["Liver protectors"]),
}
```

```python
def get_medicines(row):
    medicines = set()
    for col, (op, threshold, meds) in medicine_rules.items():
        if col in row and pd.notna(row[col]):
            try:
                value = float(row[col])
                if op == "<" and value < threshold:
                    medicines.update(meds)
            except:
                continue
    if not medicines:
        return "No medicine needed - All values normal "
    return ", ".join(sorted(medicines))
```

```python
df['Medicine'] = df.apply(get_medicines, axis=1)
```

```python
df.to_csv("updated_cbc_full.csv", index=False)
```
